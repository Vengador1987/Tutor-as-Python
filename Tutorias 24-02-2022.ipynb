```python
# Genera el primer diccionario

diccionario = dict()

for linea in datos:
    dato = linea.split(",")
    
    diccionario[random.randint(10000000000, 99999999999)] = {"nombre" : dato[1],
                                                             "atr1" : dato[2],
                                                             "atr2" : dato[3],
                                                             "total" : int(dato[4]),
                                                             "hp" : int(dato[5]),
                                                             "ataque" : int(dato[6]),
                                                             "defensa" : int(dato[7]),
                                                             "sp_atk" : int(dato[8]),
                                                             "sp_def" : int(dato[9]),
                                                             "velocidad" : int(dato[10]),
                                                             "generacion" : int(dato[11]),
                                                             "legendario" : True if dato[12].strip() == "True" else False,
                                                             "tiempo" : datetime.now()}
```

# Crear un diccionario donde las llaves sean las generaciones de los pokemons y los valores sean los indices de los pokemons y sus valores sea la informacion de pokemon.

```python
diccionario
```

```python
# Sacamos las generaciones

generaciones = list()

for llave, valor in diccionario.items():
    generaciones.append(valor["generacion"])
    
generaciones = set(generaciones)

generaciones
```

```python
# Creamos el diccionario

dicc_generaciones = dict()

for generacion in generaciones:
    
    dicc_loop = dict()
    
    for llave, valor in diccionario.items():
        if valor["generacion"] == generacion:
            dicc_loop[llave] = valor
    

    dicc_generaciones[generacion] = dicc_loop
```

```python
indice_generacion = 1
indice_pokemon = list(dicc_generaciones[1].keys())[5]

for indice_pokemon in list(dicc_generaciones[1].keys()):
    print(dicc_generaciones[indice_generacion][indice_pokemon]["nombre"])
    print(dicc_generaciones[indice_generacion][indice_pokemon]["legendario"])
    print("*"*100)
```

## Retornar una lista con el nombre y la velocidad de los pokemon donda la velocidad >= 80

```python
velocidad_pokemon = list()

for llave, valor in dicc_generaciones.items():
    indice_pokemons = list(valor.keys())
    
    for indice_p in indice_pokemons:
        if valor[indice_p]["velocidad"] >= 80:
            velocidad_pokemon.append([valor[indice_p]["nombre"],
                                      valor[indice_p]["velocidad"],
                                      valor[indice_p]["tiempo"].year])
```
